```python
import numpy as np
from PIL import Image
from scipy import misc
import os
import time
start = time.clock()

clas={1:'cottage_garden',2:'firing_range_indoor',3:'pharmacy'}
for cl in range(1,4):
    cnt=0
    base_url='./group05/test/'+clas[cl]
    jkl=clas[cl]
    print "--------------------"+jkl.upper()+"-----------------------"
    for filename in os.listdir(base_url):
        if(filename[-3:]=="jpg"):
            cnt=cnt+1
            img = Image.open(base_url+'/'+filename)
            img=img.convert('RGB')
            arr = np.array(img)
            fname=filename[:-4]

        # arr=misc.imread(filename) 
        # img = Image.open(filename)
        # arr = np.array(img)
        # arr=list(img.getdata())
        # print(len(arr), img.size , 1024*768/1024)

            width, height = img.size
            ih=height/32
            jw=width/32
        #     print ih,jw
            print cnt,filename, (width/32)*(height/32), img.size,"started.........",ih,jw

            i=0
            k=0
            j=0
            r={}
            g={}
            b={}
            for v in range(0,8):
                r[v]=0
                g[v]=0
                b[v]=0

            f = open(base_url+'/'+fname,'w+')
            for i in range(0, ih):
                for j in range(0, jw):
                    for ii in range(i*32, i*32+32):
                        for jj in range(j*32, j*32+32):
                            arr[ii][jj]=arr[ii][jj]/32
                #             print k,ii,jj,arr[ii][jj]

                            r[arr[ii][jj][0]]=r[arr[ii][jj][0]]+1    
                            g[arr[ii][jj][1]]=g[arr[ii][jj][1]]+1
                            b[arr[ii][jj][2]]=b[arr[ii][jj][2]]+1

                    k=k+1
                    vec=[r[l] for l in r]
                    vec+=[g[l] for l in g]
                    vec+=[b[l] for l in b]
                    for v in range(0,8):
                        r[v]=0
                        g[v]=0
                        b[v]=0

            #         print "Grid :","(",i,j,")",k,vec
                    f = open(base_url+'/'+fname,'a+')
                    np.savetxt(f, np.array([vec]), delimiter=' ',fmt="%d")

#                     for s in vec:
#                         f.write(str(s)+' ')
#                     f.write("\n")
                    j=j+1;
                i=i+1;
            f.close()
#your code here    
print "Time :",time.clock() - start
```
